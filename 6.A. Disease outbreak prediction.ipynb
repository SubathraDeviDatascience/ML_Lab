```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the data
disease_data=pd.read_csv('disease_outbreak.csv')
#Preview the dataset
print(disease_data.head())
```

```python
#Prepare Data
x=disease_data[['temperature','humidity','population_density']]
y=disease_data['outbreak']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Train Model
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
#Evaluate model
outbreak_predictions=model.predict(x_test)
accuracy=accuracy_score(y_test,outbreak_predictions)
print("Disease outbreak prediction accuracy:",accuracy)
```

```python
#Visualization
sns.scatterplot(data=disease_data,x='temperature',y='humidity',hue='outbreak')
plt.title("Disease outbreak prediction visualization")
plt.show()
```
