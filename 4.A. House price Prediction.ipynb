```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
#Load the dataset
df=pd.read_csv("Bouston Housing.csv")
```

```python
#Preview the dataset
print("Preview the dataset\n")
print(df.head())
```

```python
#Encode Categorical data
label_encoder=LabelEncoder()
df['Locations']=label_encoder.fit_transform(df['Locations'])
```

```python
#Separate features and target variable
x=df[['SquareFootage','Bedrooms','Bathrooms','Locations']]
y=df['Price']
```

```python
#Create and train the model
model=LinearRegression()
model.fit(x,y)
```

```python
#Make Predictions
y_pred=model.predict(x)
```

```python
# Visualization
plt.scatter(df['SquareFootage'], df['Price'], color='blue', label='Actual Price')
plt.scatter(df['SquareFootage'], y_pred, color='red', label='Predicted Price')
plt.xlabel('Square Footage')
plt.ylabel('Price')
plt.legend()
plt.title('Linear Regression on Housing Price')
plt.show()
```

```python
#Display the coefficient
print(f"Coefficients:{model.coef_}")
print(f"Interccept:{model.intercept_}")
```

```python
#Predicting a new sample
new_sample=np.array([[3500,5,4,label_encoder.transform(['Inside City'])[0]]])
predicted_price=model.predict(new_sample)
print(f"Predicted price for 3500 sqft, 5 Bedrooms, 4 Bathrooms, Inside city:{predicted_price[0]}")
```
